```python
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
ending = datetime.now()
starting = datetime(ending.year - 1, ending.month, ending.day)
#Apple
Apple = yf.download('AAPL', starting, ending)
```

```python
#New EDA - plot prices and hist of volume
#Index(['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'], dtype='object')
#print(Apple.columns)
Apple['Adj Close'].plot()
plt.show()

Apple['Volume'].plot.hist(bins=50)
plt.show()
```

```python
#Create a 10d close % change and plot a histogr
Apple['10d_adj_close_pct'] = Apple['Adj Close'].pct_change(10)
Apple['10d_adj_close_pct'].plot.hist(bins=10)
plt.show()
```
